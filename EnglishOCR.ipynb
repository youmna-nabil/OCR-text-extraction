```python
import cv2
import pytesseract
from PIL import Image
import re
import pandas as pd
import cv2
```

```python
image_path = '/content/img75.jpg'
img = cv2.imread(image_path)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

items = []
descriptions = []
qtys = []
unit_prices = []
amounts = []

pattern = r'(\d+)\s+\d+\s+([A-Za-z0-9\s\-\(\)#]+)\s+(\d+)\s+(\d+\.\d{2})\s+(\d+\.\d{2})'

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)

    roi = img[y:y+h, x:x+w]
    roi_pil = Image.fromarray(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))

    text = pytesseract.image_to_string(roi_pil, config='--psm 6')

    match = re.search(pattern, text)
    if match:
        items.append(match.group(1))
        descriptions.append(match.group(2))
        qtys.append(match.group(3))
        unit_prices.append(match.group(4))
        amounts.append(match.group(5))

data = {
    "Item": items,
    "Description": descriptions,
    "QTY": qtys,
    "UNIT_PRICE": unit_prices,
    "Amount": amounts
}

df = pd.DataFrame(data)
print(df)
```

```python
image_path = '/content/img75.jpg'
img = cv2.imread(image_path)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY_INV)

contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

all_text = ''

for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)

    roi = img[y:y+h, x:x+w]
    roi_pil = Image.fromarray(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    text = pytesseract.image_to_string(roi_pil, config='--psm 6')
    all_text += text + '\n'

pattern = r'(\d+)\s+\d+\s+([A-Za-z0-9\s\-\(\)#]+)\s+(\d+)\s+(\d+\.\d{2})\s+(\d+\.\d{2})'

items = []
descriptions = []
qtys = []
unit_prices = []
amounts = []

matches = re.findall(pattern, all_text)
for match in matches:
    items.append(match[0])
    descriptions.append(match[1])
    qtys.append(match[2])
    unit_prices.append(match[3])
    amounts.append(match[4])


data = {
    "Item": items,
    "Description": descriptions,
    "QTY": qtys,
    "UNIT_PRICE": unit_prices,
    "Amount": amounts
}

df = pd.DataFrame(data)
print(df)
```

```python
table_text = []
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    roi = img[y:y+h, x:x+w]
    roi_pil = Image.fromarray(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
    text = pytesseract.image_to_string(roi_pil, config='--psm 6')
    table_text.append(text.strip())

for text in table_text:
    print(text)
```

```python
from langchain.chains import LLMChain
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from google.colab import userdata
from langchain.prompts import PromptTemplate
import csv
```

```python
reader = easyocr.Reader(['ar', 'en'])
result = reader.readtext('/content/img75.jpg')

for detection in result:
    text = detection[1]
```

```python
all_text = ""
for detection in result:
    text = detection[1]
    all_text += text + '\n'

print(f"Extracted Text:\n{all_text}")
```

```python
google_api_key = userdata.get('google_api_key')
llm = ChatGoogleGenerativeAI(google_api_key=google_api_key,  temperature=0, model="gemini-1.5-flash")
```

```python
prompt_template = PromptTemplate(
    input_variables=["extracted_text"],
    template="""Based on the following text extracted from an image:\n\n{extracted_text}\n\n
I want you to extract the following information:\n
1. Item (Part Number)\n
2. Description\n
3. Quantity (QTY)\n
4. Unit Price\n
5. Amount\n
Please provide the extracted data in a structured table format that can easily be converted into a CSV file.
The format should look like this:\n
Item,Description,QTY,Unit Price,Amount\n
1,Sample Item 1,10,2.50,25.00\n
2,Sample Item 2,5,3.00,15.00\n
... (and so on)\n
Make sure to separate each value with a comma and provide a new line for each item.
"""
)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template)
response = chain.run({"extracted_text": all_text})

print(response)
```

```python
# Input data in the specified format
data = response

# Splitting the data by lines and then by commas
lines = data.strip().split('\n')
data_rows = [line.split(',') for line in lines]

# Saving the data into a CSV file
csv_file_path = 'output_data.csv'
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data_rows)

print(f"Data saved to {csv_file_path} successfully.")
```
